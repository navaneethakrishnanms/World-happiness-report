part 3-Statistics

```python
mean_hp=df['Ladder score'].mean()
print("Mean:",mean_hp)
median=df['Ladder score'].median()
print("Median:",median)
var=df['Ladder score'].var()
print("Variance:",var)
std=df['Ladder score'].std()
print("ST deviation:",std)

min=df['Ladder score'].min()
print("Min score:",min)
max=df['Ladder score'].max()
print("Max score:",max)
mode=df['Ladder score'].mode()
print("Mode:",mode)
```

```python

global_avg = df['Ladder score'].mean()

below_avg_count = df[df['Ladder score'] < global_avg]['Country name'].nunique()


total_countries = df['Country name'].nunique()


percentage_below_avg = (below_avg_count / total_countries) * 100

print(f"Global average happiness score: {global_avg:.2f}")
print("Total countries:",total_countries)
print(f"Number of countries below average: {below_avg_count}")
print(f"Percentage of countries below average: {percentage_below_avg:.2f}%")
```

part 4-linear algebra

```python
ladder_score=df['Ladder score'].to_numpy()
gdp=df['Log GDP per capita'].to_numpy()
print("Vector addition:",ladder_score+gdp)
print("Vector subtraction:",ladder_score-gdp)
dot_product=np.dot(ladder_score,gdp)
print("Dot product:",dot_product)



```

```python


matrix = df[['Log GDP per capita', 'Social support']].to_numpy()
weight=np.array([0.5,0.5])
print("Weighted matrix multiplication:",np.dot(matrix,weight))
```

```python
ladder_score=df['Ladder score'].to_numpy()
ladder_score_normalized = (ladder_score - ladder_score.min()) / \
                          (ladder_score.max() - ladder_score.min())

print("First 5 normalized values:", ladder_score_normalized)
```

Part 5 Calculus

```python
import sympy as sp
GDP,ladder_score=sp.symbols('GDP ladder_score')
Wellbeing_Index = (GDP*ladder_score) + 0.5*(ladder_score-5)**2
derivative=sp.diff(Wellbeing_Index,ladder_score)
print("Derivative:",derivative)
```
